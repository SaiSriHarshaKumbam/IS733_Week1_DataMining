```python
#reading the dataset into a dataframe 
#df = pd.read_csv('C:/Users/harsha/Desktop/DataMining_733/DataAnalyst.csv')

# Clean the "Salary Estimate" column to keep only numeric values.
#df['Cleaned Salary'] = df['Salary Estimate'].str.replace('[^\d]', '', regex=True)

# Display the DataFrame with the cleaned "Cleaned Salary" column.
#print(df[['Cleaned Salary']])

# Sample DataFrame with the "Salary Estimate" column.
#data = {'Salary Estimate': ['$37K-$66K (Glassdoor est.)', '$46K-$87K (Glassdoor est.)']}
#df = pd.DataFrame(data)
#Question-2
df = pd.read_csv('C:/Users/harsha/Desktop/DataMining_733/DataAnalyst.csv')
# Clean the "Salary Estimate" column to keep only numeric values.
df['Cleaned Salary'] = df['Salary Estimate'].str.replace('[^\d]', '', regex=True)
#print('Cleaned Salary is as follows: ')
#print(df)

# Split the +"Salary Estimate" column into low and high salary columns.
split_salaries = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')
#print('Split Salaries are as follows: ')
#print(split_salaries)
# Rename the columns for clarity.
split_salaries.columns = ['Cleaned_Low_Salary', 'Cleaned_High_Salary']
print(split_salaries.columns)
# Combine the cleaned salary columns with the original DataFrame.
df = pd.concat([df, split_salaries], axis=1)

# Display the DataFrame with the cleaned columns.
print(df[['Cleaned Salary', 'Cleaned_Low_Salary', 'Cleaned_High_Salary']])


# Convert the "Cleaned Low Salary" and "Cleaned High Salary" columns to numeric.
df['Cleaned_Low_Salary'] = pd.to_numeric(split_salaries['Cleaned_Low_Salary'])
df['Cleaned_High_Salary'] = pd.to_numeric(split_salaries['Cleaned_High_Salary'])

# Calculate the mid salary.
df['Mid_Salary'] = (df['Cleaned_Low_Salary'] + df['Cleaned_High_Salary']) / 2

# Display the DataFrame with the cleaned and calculated columns.
print(df[['Cleaned Salary', 'Cleaned_Low_Salary', 'Cleaned_High_Salary', 'Mid_Salary']])
```

```python
# Calculate the mean salary for the top 5 industries.
mean_salary = df[df['Industry'].isin(industry_counts.index)].groupby('Industry')['Mid_Salary'].mean().reset_index()

# Display the mean salary for the top 5 industries.
print(mean_salary)
```

```python
#Question-3
#In this question I was curious about how the type of ownerships is impacting the industries, so what I did is I took the total count of various type of ownerships and after that I have removed the Unknown or missing values in the dataset 
#Later I got a count of various type of ownerships among these ownerships I took the top 3 most frequent ownerships and created a bar plot of the top 3 ownerships with the top 5 industries that we have analyzed
# Filter the DataFrame to include only the rows with the top 5 industries (excluding '-1').
top_5_industries = industry_counts.index
filtered_df = df[df['Industry'].isin(top_5_industries)]

# Get the unique values and counts of "Type of ownership" for the top 5 industries.
ownership_counts = filtered_df.groupby(['Industry', 'Type of ownership']).size().reset_index(name='Count')

# Display the unique values and counts.
print(ownership_counts)
```

```python
# Removing the Unknown values or missing values from the type of ownership
# Filter the DataFrame to include only the rows with the top 5 industries (excluding '-1').
top_5_industries = industry_counts.index
filtered_df = df[df['Industry'].isin(top_5_industries)]

# Filter out rows where 'Type of ownership' is 'Unknown'.
filtered_df = filtered_df[filtered_df['Type of ownership'] != 'Unknown']

# Get the unique values and counts of "Type of ownership" for the top 5 industries.
ownership_counts = filtered_df.groupby(['Industry', 'Type of ownership']).size().reset_index(name='Count')

# Display the unique values and counts.
print(ownership_counts)
```

```python
#Question-3
# Get the top 5 industries.
top_5_industries = industry_counts.index

# Create an empty dictionary to store the top 3 ownership counts for each industry.
top_ownership_counts = {}

# Iterate through the top 5 industries and get the top 3 ownership counts for each.
for industry in top_5_industries:
    industry_data = ownership_counts[ownership_counts['Industry'] == industry]
    top_3_ownership = industry_data.nlargest(3, 'Count')
    top_ownership_counts[industry] = top_3_ownership

# Create subplots for each industry.
fig, axes = plt.subplots(nrows=1, ncols=len(top_5_industries), figsize=(25, 20))

# Iterate through the top 5 industries and create bar plots for each.
for i, industry in enumerate(top_5_industries):
    ax = axes[i]
    ownership_data = top_ownership_counts[industry]
    ax.bar(ownership_data['Type of ownership'], ownership_data['Count'], color='skyblue')
    ax.set_title(f'Top 3 Ownership Types in {industry}')
    ax.set_xlabel('Type of Ownership')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=45)
    
plt.tight_layout()
plt.show()
```
